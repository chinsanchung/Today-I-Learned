```python
import matplotlib.pyplot as plt
```

```python
# plot(): 직선, 꺾은선 그래프..인수는 Y축값, X축은 자동으로 0부터 1씩 증가
plt.plot([10,20,30,40])
plt.show()
```

```python
# 첫 인수가 X축, 두번째가 Y축
plt.plot([1,2,3,4], [12,43,25,15])
plt.show()
```

```python
# 제목, 범례 넣기...색상, 선 모양 변경
plt.title('test')
plt.plot([10,20,30,40], color='green', linestyle='--', label='asc') # 줄이면 plt.plot([1,2], 'g--')
plt.plot([40,30,20,10], color='blue', ls=':', label='desc') # 줄이면 plt.plot([1,2,3,4], 'b:')
## legend()로 범례 실행
plt.legend()
plt.show()
```

```python
# 마커 모양 바꾸기
plt.title('marker')
plt.plot([10,20,30,40], 'r', label='circle')
plt.plot([40,30,20,10], 'g^', label='triangle up')
plt.legend()
plt.show()
```
